```python
import pandas as pd

def ler_planilha(caminho):
    df = pd.read_excel(
        caminho,
        sheet_name=1,
        names=['conta contabil', 'valor', 'classificacao'],
        skiprows=2
    )

    df['valor'] = pd.to_numeric(df['valor'], errors='coerce').fillna(0)

    return df.to_dict(orient='records')

def classificar_conta(contas):
    balanco = { #Dicionário das contas
        "receita" : [],
        "despesa" : [],
        "custos" : [],
        "liquidas": []
    }
    palavras_passivo = ["despesa", "custo"]
    palavras_ativo = ["receita"]
    palavras_custo = ["DAS", "Abatimentos"]
    palavras_liquidas = ["CSP", "CMV"]

    for conta in contas:
      nome = str(conta["classificacao"]).lower()

      if any(p in nome for p in palavras_passivo):
          conta["classificacao"] = "despesa"
          balanco['despesa'].append(conta)

      elif any(p in nome for p in palavras_ativo):
          conta["classificacao"] = "receita"
          balanco['receita'].append(conta)

    for conta in contas:
      nome = str(conta["conta contabil"])

      if any(p in nome for p in palavras_custo):
          balanco['custos'].append(conta)

      elif any(p in nome for p in palavras_liquidas):
          balanco['liquidas'].append(conta)

    return balanco

def somar_balanco(balanco_total):
    total_custos = sum(item["valor"] for item in balanco_total['custos'])
    total_liquidas = sum(item["valor"] for item in balanco_total['liquidas'])
    total_ativos = sum(item["valor"] for item in balanco_total['receita'])
    total_passivos = sum(item["valor"] for item in balanco_total['despesa'])

    patrimonio_liquido = total_ativos - total_passivos
    print(f"Balanço: ")
    print(f"Total Ativo:  R$ {total_ativos:,.2f}")
    print(f"Total Passivo: R$ {total_passivos:,.2f}")
    print(f"Total Custos:  R$ {total_custos:,.2f}")
    print(f"Total Liquidas: R$ {total_liquidas:,.2f}")

    print(f"Receita Liquida: R$ {((total_ativos - 1500) + total_custos):,.2f}")
    print(f"Margem de contribuição: R$ {(((total_ativos - 1500) + total_custos) + total_liquidas):,.2f}")
    print(f"Resultado operacional: R$ {(total_ativos + total_passivos):,.2f}")
    resultado_operacional = total_ativos + total_passivos
    porcentagem = (resultado_operacional / total_ativos) * 100
    print(f"Porcentagem do total de Ativos: {porcentagem:.2f}%")

def main():
  caminho = "Contas.xlsx"
  lista_bruta = ler_planilha(caminho)
  balanco_final = classificar_conta(lista_bruta)
  print(f"Contas:")
  for item in lista_bruta:
    print(f"{item["conta contabil"], item["valor"], item["classificacao"]}")
  somar_balanco(balanco_final)
if __name__ == "__main__":
   main()
```
