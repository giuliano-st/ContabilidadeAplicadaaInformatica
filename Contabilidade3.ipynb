```python
import pandas as pd

def ler_planilha(caminho):
        df = pd.read_excel(caminho, names=["gasto", "valor", "classificacao"])
        df["valor"] = df["valor"].astype(str)
        df["valor"] = (
            df["valor"].str.replace("R$", "", regex=False).str.replace(" por unidade", "", regex=False).str.strip())
        df["valor"] = df["valor"].str.replace(".", "", regex=False)
        df["valor"] = df["valor"].str.replace(",", ".", regex=False)
        df["valor"] = pd.to_numeric(df["valor"], errors="coerce").fillna(0)
        return df.to_dict(orient="records")

def classificar_conta(contas):
    balanco = {
        "variável" : [],
        "fixo" : [],
        "despesa" : []
    }
    palavras_fixo = ["fixo"]
    palavras_variavel = ["variável"]
    palavras_despesa = ["despesa"]

    for conta in contas:
      nome = str(conta["classificacao"]).lower()

      if any(p in nome for p in palavras_fixo):
          balanco['fixo'].append(conta)

      elif any(p in nome for p in palavras_variavel):
          balanco['variável'].append(conta)

      elif any(p in nome for p in palavras_despesa):
                  balanco['despesa'].append(conta)

    return balanco

def somar_balanco(balanco_total):
    preco_venda = 25
    total_fixos = sum(item["valor"] for item in balanco_total['fixo'])
    total_variavel = sum(item["valor"] for item in balanco_total['variável'])
    total_despesa = sum(item["valor"] for item in balanco_total['despesa'])
    cvt = total_variavel * 4000
    cfu = total_fixos / 4000
    custo_total = total_variavel + cfu
    mc = preco_venda - custo_total - 1
    pec = total_fixos / mc

    print(f"Dados: ")
    print(f"Custo Variável Unitario:  R$ {total_variavel:,.2f}")
    print(f"Custo Variável Total: R$ {cvt}")
    print(f"Custo Fixo Unitario:  R$ {cfu}")
    print(f"Custo Fixo Total:  R$ {total_fixos:,.2f}")
    print(f"Custo Total: R$ {custo_total:,.2f}")
    print(f"Margem de Constribuição Unitária: R$ {mc:,.2f}")
    print(f"Ponto de Equilibrio Contábil: {pec:,.2f} potes")

def main():
  caminho = "Dados Custos.xlsx"
  lista_bruta = ler_planilha(caminho)
  balanco_final = classificar_conta(lista_bruta)
  print(f"Contas:")
  for item in lista_bruta:
    print(f"{item["gasto"], item["valor"], item["classificacao"]}")
  somar_balanco(balanco_final)
if __name__ == "__main__":
   main()
```
